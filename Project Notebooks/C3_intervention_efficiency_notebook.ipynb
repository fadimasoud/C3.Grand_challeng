**The power of mask mandates**


Masks is an efficient and cost-effective way to help prevent the spread of the coronavirus.

Take Kansas, [where a real-world experiment in face coverings emerged this summer.](https://nl.nytimes.com/f/a/92ZxrtdyW_DzyXbJrEEQeA~~/AAAAAQA~/RgRhecxGP4Q_AWh0dHBzOi8vd3d3Lm55dGltZXMuY29tL2xpdmUvMjAyMC8xMC8yNi93b3JsZC9jb3ZpZC0xOS1jb3JvbmF2aXJ1cy11cGRhdGVzL2thbnNhcy1jb3VudGllcy13aXRoLW1hc2stbWFuZGF0ZXMtaGFkLWhhbGYtdGhlLXJhdGUtb2YtbmV3LWluZmVjdGlvbnMtYS1zdHVkeS1maW5kcz9jYW1wYWlnbl9pZD0xNTQmZW1jPWVkaXRfY2JfMjAyMDEwMjYmaW5zdGFuY2VfaWQ9MjM1MDkmbmw9Y29yb25hdmlydXMtYnJpZWZpbmcmcmVnaV9pZD0xMTI5MTUxODImc2VnbWVudF9pZD00MjM1NSZ0ZT0xJnVzZXJfaWQ9ZTAyOWNhOTYxMjA5MjY1YTQ5ZmUxODhjYTg1ODI2NmFXA255dEIKAChGR5dfgTCYHVIYZmFkaW1hc291ZDExZmVAZ21haWwuY29tWAQAAAAA) In early July, Gov. Laura Kelly, a Democrat, issued a statewide mask order, but was forced to let counties opt out of it under a law limiting her emergency management powers.

Only 20 of the state’s 105 counties enforced the order, which required residents to wear masks in public. Those 20 counties saw half as many new coronavirus infections as the counties that did not have the mandate in place, according to [a new study from the University of Kansas](https://nl.nytimes.com/f/a/Kucr-VigkARkDyO97O_OCw~~/AAAAAQA~/RgRhecxGP0TOaHR0cHM6Ly9tZWRpYWh1Yi5rdS5lZHUvbWVkaWEvTWFza3MvMV80OWJiOWFpZD9jYW1wYWlnbl9pZD0xNTQmZW1jPWVkaXRfY2JfMjAyMDEwMjYmaW5zdGFuY2VfaWQ9MjM1MDkmbmw9Y29yb25hdmlydXMtYnJpZWZpbmcmcmVnaV9pZD0xMTI5MTUxODImc2VnbWVudF9pZD00MjM1NSZ0ZT0xJnVzZXJfaWQ9ZTAyOWNhOTYxMjA5MjY1YTQ5ZmUxODhjYTg1ODI2NmFXA255dEIKAChGR5dfgTCYHVIYZmFkaW1hc291ZDExZmVAZ21haWwuY29tWAQAAAAA).

Cellphone-tracking data from the University of Maryland showed no differences in how often people left home in the counties with or without mask mandates, so it seemed likely that the masks made the difference.

Experts say it’s part of a countrywide trend: Localities that impose mask mandates often see fewer cases, fewer hospitalizations, fewer deaths and lower test-positivity rates than nearby localities that do not.

Other studies have turned up similar results in Alabama, Oklahoma, South Carolina and Texas. A recently published report from the Centers for Disease Control and Prevention found [a 75 percent drop in coronavirus cases in Arizona](https://nl.nytimes.com/f/a/dVJWiXrz3vbMFKfbWYukXA~~/AAAAAQA~/RgRhecxGP0TTaHR0cHM6Ly93d3cuY2RjLmdvdi9tbXdyL3ZvbHVtZXMvNjkvd3IvbW02OTQwZTMuaHRtP2NhbXBhaWduX2lkPTE1NCZlbWM9ZWRpdF9jYl8yMDIwMTAyNiZpbnN0YW5jZV9pZD0yMzUwOSZubD1jb3JvbmF2aXJ1cy1icmllZmluZyZyZWdpX2lkPTExMjkxNTE4MiZzZWdtZW50X2lkPTQyMzU1JnRlPTEmdXNlcl9pZD1lMDI5Y2E5NjEyMDkyNjVhNDlmZTE4OGNhODU4MjY2YVcDbnl0QgoAKEZHl1-BMJgdUhhmYWRpbWFzb3VkMTFmZUBnbWFpbC5jb21YBAAAAAA~) less than a month after mask-wearing became enforced and bars and gyms were shuttered.

```
#@title
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from IPython.display import Javascript
```

```
#@title
df = pd.read_csv('/content/all-states-history.csv')
df.fillna(0, inplace=True)
df['date'] = pd.to_datetime(df['date'])
# create new column for daily Increase Positive Cases Percenatage Per Total daily increase in Total Tests Result (%*100)
df['positiveIncreasePerTestIncrease']=((df['positiveIncrease']+1)/(df['totalTestResultsIncrease']+1))*10000
# print('''02_COVID_Tracking_Project CSV data definitions can be found on: 
# https://covidtracking.com/about-data/data-definitions''')
```
